# Read a txt file

Load modules

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Read a .txt file with temperatures. The file is on a GitHub repository (https://github.com/hau-mech/test_ReadTXT.git)

```python
file_path = "https://raw.githubusercontent.com/hau-mech/test_ReadTXT/main/Temperature.txt"
df = pd.read_csv(file_path, sep = " ", engine = "python", header = None)
print(df)
```

Add headers

```python
header_text = ['date', 'time', 'temperature']
df.columns = header_text
print(df)
```

Create a column with dates and times, and convert it to "datetime" format

```python
df["date_time"] = df["date"] + " " + df["time"]
df["date_time"] = pd.to_datetime(df['date_time'])
print(df)
```

Plot temprature vs. time

```python
# Create a plot
plt.figure(figsize = (12, 6))  # Adjust figure size if needed
plt.plot(df.date_time, df.temperature, marker = 'o', linestyle = '-', markersize = 5)
plt.title('Temperature Measurements Over Time')
plt.xlabel('Date and Time')
plt.ylabel('Temperature (Celsius)')
plt.xticks(rotation = 45)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
